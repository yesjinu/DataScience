```python
import pandas as pd

soogang = pd.read_csv('data/enrolment_1.csv')
soogang
```

### DataFrame에 status 칼럼 추가

```python
soogang['status'] = 'allowed'
soogang
```

### 조건 1과 조건 2를 만족시키기 위한 조건문
조건 1 = 1학년과 imformation technology 과목 같이 못 함  
조건 2 = 4학년과 commerce 과목 같이 못 함

```python
con1 = (soogang['year'] == 1) & (soogang['course name'] == 'information technology') 
con2 = (soogang['year'] == 4) & (soogang['course name'] == 'commerce')
condition = con1 | con2
soogang.loc[condition, 'status'] = 'not allowed'
condition
```

### 조건 3을 만족시키기 위한 조건문
조건 3 = 수강생 5명 이하 과목은 폐강

```python
allowed_courses_yet = soogang['status'] == 'allowed'
closed_courses = soogang.loc[allowed_courses_yet, 'course name'].value_counts()
sample = list(closed_courses[closed_courses < 6].index)
```

```python
for course in sample:
    soogang.loc[soogang["course name"] == course, 'status'] = 'not allowed'
```
